```python
import pandas as pd
from sklearn import preprocessing
```

```python
data = pd.read_csv("../outputs/clean_houses.csv")
```

```python
data.rename(columns={"Unnamed: 0":"index"},inplace=True)
data.set_index("index", inplace=True)
data.head()
```

```python
def transf(column):

    le = preprocessing.LabelEncoder()
    data[column] = le.fit_transform(data[column])
    return data[column]
```

```python
transf("is_renewal_needed")
transf("is_new_development")
transf("is_exterior")
transf("has_lift")
transf("has_parking")
transf("has_storage_room")
transf("has_terrace")
transf("has_balcony")
transf("has_pool")
```

```python
data.sq_mt_built.value_counts().head(300)
```

```python
data = data.drop(data[data["buy_price"]>1200000].index)
data = data.drop(data[data["sq_mt_built"]>300].index)
data = data.drop(data[data["n_rooms"]>5].index)
data = data.drop(data[data["n_bathrooms"]>5].index)
```

```python
data.reset_index()
data
```

```python
# Aplicamos ahora get_dummies a la columna "house_type_id" y "floor".

data = pd.get_dummies(data, columns = ["house_type_id","floor"])
```

```python
data.columns
```

```python
data = data[['is_renewal_needed', 'is_new_development',
       'sq_mt_built', 'n_rooms', 'n_bathrooms', 'is_exterior', 'has_lift',
       'has_parking', 'has_storage_room', 'has_terrace', 'has_balcony',
       'has_pool', 'latitude', 'longitude',
       'house_type_id_HouseType 1: Pisos', 'house_type_id_HouseType 4: Dúplex',
       'house_type_id_HouseType 5: Áticos', 'floor_1', 'floor_2', 'floor_3',
       'floor_4', 'floor_5', 'floor_6', 'floor_7', 'floor_8', 'floor_9',
       'floor_Bajo', 'floor_Entreplanta exterior',
       'floor_Entreplanta interior', 'floor_Semi-sótano exterior',
       'floor_Semi-sótano interior', 'floor_Sótano exterior',
       'floor_Sótano interior', 'buy_price']]
```

### Iniciamos el análisis con H2O

```python
import h2o
from h2o.estimators import H2ORandomForestEstimator

h2o.init()
```

```python
h2train = h2o.H2OFrame(data)
```

```python
columnas = [col for col in data.columns if col not in ["buy_price"]]

x = list(columnas)

y = "buy_price"
```

```python
from h2o.automl import H2OAutoML

automl = H2OAutoML(max_runtime_secs=1200, sort_metric='RMSE')
automl.train(x, y, training_frame=h2train)
leader_board = automl.leaderboard
leader_board.head()
```

```python
from h2o.automl import H2OAutoML

automl = H2OAutoML(max_models=50, seed=1,max_runtime_secs=1800, sort_metric='RMSE')
automl.train(x, y, training_frame=h2train)
```

```python
leader_board = automl.leaderboard
leader_board.head()
```
